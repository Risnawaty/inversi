```python
import matplotlib.pyplot as plt
import numpy as np
from numpy.linalg import inv
```

```python
#Data Observasi
t = np.array ([0.00, 0.25, 0.50, 0.75, 1.00, 1.25, 1.50, 1.75, 2.00, 2.25, 2.50, 2.75, 3.00, 3.25, 3.50, 3.75, 4.00, 4.25, 4.50, 4.75, 5.00])
h = np.array ([5.10, 5.10, 6.10, 6.10, 7.10, 7.10, 7.10, 8.10, 8.10, 8.10, 7.10, 7.10, 7.10, 6.10, 6.10, 5.10, 4.10, 3.10, 2.10, 1.10, 0.10])
h0 = 5
```

```python
#Membuat Matriks G
G = np.zeros((len(t), 2))

for i in range(len(t)):
    G[i,0] = t[i]
    G[i,1] = t[i]**2
```

```python
#Membuat Matriks d
d = np.reshape(h - h0, (len(h), 1))
print('d=\n',d)
```

```python
GT = G.T
GTG = GT @ G
GTd = GT @ d

m = np.linalg.inv(GTG) @ GTd
```

```python

v0 = m[0,0]
g = -2 * m[1,0]

print("Kecepatan awal (v0) =", v0, "m/s")
print("Gravitasi planet (g) =", g, "m/s^2")
```
